```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generování 1000 náhodných bodů
np.random.seed(10)
x = np.random.rand(500) * 100
y = np.random.rand(500) * 100
z = np.random.rand(500) * 100

# Vytvoření 3D grafu
fig = plt.figure(figsize=(4,5))
ax = fig.add_subplot(111, projection='3d')

# Zobrazení bodů v grafu
ax.scatter(x, y, z, c='blue', marker='o', s=10)

# Nastavení popisků os
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Nastavení limity os
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_zlim(0, 100)



plt.savefig("randompoints.pdf", bbox_inches='tight')
```

```python
center1 = (50, 50, 50)
distance = 15

x1 = np.random.uniform(center1[0], center1[0] + distance, size=(500,))
y1 = np.random.normal(center1[1], distance, size=(500,))
z1 = np.random.normal(center1[2], distance, size=(500))

fig = plt.figure(figsize=(4,5))
ax = fig.add_subplot(111, projection='3d')

# Zobrazení bodů v grafu
ax.scatter(x1, y1, z1, c='blue', marker='o', s=10)

# Nastavení popisků os
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Nastavení limity os
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_zlim(0, 100)

plt.savefig("clusterpoints.pdf", bbox_inches='tight')
```
