```python
import pyvoro
import matplotlib.pyplot as plt
import numpy as np

# Generujeme náhodná data pro body
np.random.seed(42)
points = np.random.rand(20, 2)  # 20 bodů ve 2D prostoru

# Definujeme funkci pro vytvoření Vornoiho diagramu
def compute_voronoi(points):
    # Vytvoříme objekt VornoiDiagram
    voronoi_diagram = pyvoro.compute_voronoi(points.tolist(), [[0.0, 1.0], [0.0, 1.0]])

    # Získáme informace o hranách Vornoiho diagramu
    edges = []
    for region in voronoi_diagram['regions']:
        if -1 not in region and len(region) > 0:
            edges.extend(zip(region, region[1:] + [region[0]]))

    return edges

# Vypočítáme Vornoiho diagram
edges = compute_voronoi(points)

# Vykreslíme body a hrany Vornoiho diagramu
plt.scatter(points[:, 0], points[:, 1], c='red', marker='o')
for edge in edges:
    plt.plot(*zip(*points[edge]), color='blue')

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
